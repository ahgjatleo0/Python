### 산술 연산자

#### 숫자 연산

```python
num1 = 10
num2 = 7

# 나누기 / , 나머지 % , 몫 //

a = num1/num2
b = num1%num2
c = num1//num2
print(f'나누기 : ',a)
print(f'나머지 : ',b)
print(f'몫 : ',c)
```

#### 문자열 더하기

```python
str1 = '안녕'
str2 = '하세요!'
# 문자열 +
str1 + str2
```

```python
str1 = '10'
str2 = '7'
# 숫자를 문자열로

str1 + str2
```

```python
# 문자열을 체크하는 함수 : isalpha(), isdigit()
# 문자열 전처리(깔끔하게 정리하는 작업) 할 때 사용 ==> 조건문에서 활용

# isalpha() : 문자열이 문자인지 아닌지를 확인하는 함수
# 숫자가 단 하나라도 있다면 ==> False 반환
# 한글도 문자로 인식 ==> True 반환

# isdigit() : 문자열이 숫자인지 아닌지를 확인하는 함수
# 문자가 단 하나라도 있다면 ==> False 반환
# 음수부호(-), 소수점(.)을 문자로 판단 ==> 실수나 음수를 판단하지 못함(False 반환)
# 오직 0을 포함한 양수형 정수, 지수 ==> True 반환

str1.isdigit()
str1.isalpha()
```

### 비교 연산자

```python
# True, False로 결과가 출력되는 연산자
# > , >= , < , <= , ==(같다) , !=(같지 않다)

a = 3
b = 7

print(a > b)
print(a <= b)
print(a == b)
print(a != b)
```

### 논리 연산자

```python
# not, and, or
# not : 논리를 부정하는 연산자 True -> False, False -> True
a,b = 5,10
not a < b
```

```python
# and : 두 개의 논리를 연결하여 확인, 두 논리가 모두 참일 경우에만 True
# or : 한 개의 논리만 참일 경우에도 True

print(3 > 5 and 10 == 10)
print(3 > 5 or 10 == 10)
```

### 삼항 연산자
- (True 일때 실행문장) if 조건식 else (False 일때 실행문장)

```python
# 60점 이상이면 합격, 미만이면 불합격 출력해보기
score = 50
'합격' if score >= 60 else '불합격'
```

```python
a = int(input('정수입력 >> '))
b = int(input('정수입력 >> '))

print(('a : {}'.format(a) )if a > b else ('b : {}'.format(b)))
```

```python
a = int(input('정수입력 >> '))
b = int(input('정수입력 >> '))
c = a-b if a > b else b-a
print('두 수의 차 : {}'.format(c))
```

```python
num = int(input('정수입력 >> '))
result = '짝수입니다.' if num % 2 == 0 else '홀수입니다.'
print('{}는(은) {}'.format(num, result))

```
