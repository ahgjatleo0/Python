### 문자열 함수

```python
# day 문자열에서 '맑음' 인덱스 번호 확인하기
# find 함수 활용
a = day.find('맑음')
day[a:a+2]
```

```python
# find 함수를 활용하여 날짜 인덱스 번호 확인하기
# 끝 인덱스를 포함하기 위해 + 1
b = day.find('일')
day[:b+1]
```

```python
# index 함수
day.index('일')
```

```python
# index, find 함수 차이 비교 : 문장안에서 없는 문자열을 찾을 때
day.index('e') # 문자를 찾을 수 없다면 에러 발생
day.find('e')  # 찾는 문자가 없을 때 -1을 반환, 에러를 발생하지 않는 상황으로 사용 
```

```python
# len() : 문자열 개수 반환(길이 반환)
len(day) # 공백, 특수기호(! ? .)등을 모두 포함한 길이를 반환
```

```python
# replace('찾을 문자열', '변경할 문자열')
# day : 맑음 --> 흐림
day = day.replace('맑음', '흐림')
day
```

```python
# strip() : 양쪽 공백을 제거해 주는 함수
day2 = '  우리는 빅데이터 2교육실 입니다!!! '
day2.strip()
```

```python
# lstrip() : 왼쪽 공백을 제거하는 함수
# rstrip() : 오른쪽 공백을 제거하는 ㅎ마수
day2.lstrip()
day2.rstrip()
```

```python
# count('찾을 문자열') : '찾을 문자열'이 몇 개 있는지 개수 반환하는 함수
test = 'this subject is python!!!'
test.count('is')
test.count('s')
```

```python
# 대문자 --> 소문자 : lower
# 소문자 --> 대문자 : upper
test.upper()
```

```python
# split() : 문자열 나눠주는 함수
test.split() # 공백을 기준으로 문자열을 나눔
```

```python
# 특정 기호를 기준으로 문자열을 나누고 싶을 때
s2 = '오늘은. 즐거운. 월요일입니다.'
s2.split('.')
```

### 문자열 포매팅
- % 기호 포매팅
- format 함수를 사용한 포매팅
- f문자열 포매팅

```python
# '%d'를 이용해서 변수를 대입해보기
x = 14
s = '내일은 3월 %d일 입니다.' %x
s
```

```python
month = 3
day = 13
s = '오늘은 %d월 %d일 입니다.'%(month,day)
s
```

#### format 함수를 사용한 포매팅

```python
month = 3
day = 13
s = '오늘은 {}월 {}일 입니다.'.format(month, day)
s
```

```python
# 내일은 3월 14일입니다.
s = '내일은 3월 {}일입니다.'.format(14)
s
```

```python
# 인덱스 개념을 이용한 format 함수 포매팅
s = '오늘은 {0}월 {1}일 {1}요일 입니다.'.format(3,13)
s
```

```python
month = 3
day = 13
s = f'오늘은 {month}월 {day}일 입니다.'
s
```
